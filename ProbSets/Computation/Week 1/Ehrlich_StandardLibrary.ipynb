```python
"""
Problem 1
"""

def prob1(l):
    '''
    Function takes a list of numbers and returns the maximum, minimum and average. 
    
    Input: List
    Output: Maximum (int), Minimum (int), Average (float)
    '''
    return max(l), min(l), float(sum(l))/len(l)

l = [4,5,6,7,8]
prob1(l)
```

```python
"""
Problem 2
"""

def prob2(obj1):
    """
    Function checks if object is immutable. Returns boolean.
    """
    obj2= obj1
    if type(obj1) is list:
        obj2.append(1)
    elif type(obj1) is dict:
        obj2[1] = 'a'
    elif type(obj1) is tuple:
        obj2 += (1,)
    elif type(obj1) is str:
        obj2 += 'a'
    else:
        obj2+=1
        
    Bool = obj2 == obj1
    return print(str(type(obj1)) +  " is mutable: " + str(Bool))
        
my_l = []
my_dict = {}
my_num = 0
my_str = ''
my_tuple = ()

prob2(my_l)
prob2(my_dict)
prob2(my_str)
prob2(my_num)
prob2(my_tuple)
```

```python
"""
Problem 3: 

Create module that calculates the sum and product of two arguments, and a square root of a single argument. 
Use the module to calculate the hypotenuse of a right triangle given other two sides.
"""
import Calculator as calc

def prob3(a, b):
    '''
    Calculate the hypotenuse of a right triangle given two sides.
    
    Input:
    Side 1 - number (either int or float)
    Side 2 - number (either int or float)
    
    Output: 
    Hypotenuse - number
    
    '''
    hypotenuse  = calc.sqrt(calc.sum(calc.prod(a,a), calc.prod(b,b)))
    return hypotenuse 

side1 = 3
side2 = 4
prob3(side1, side2)
```

```python
"""
Problem 4
"""
import sys
import box
import random

def get_name():
    '''
    Asks for users name. If no name is entered after command line,
    function prompots user for name. 
    
    Output: Player's name as string
    
    '''
    if len(sys.argv) != 2:
        name = input("What is your name? ")
    else:
        name = sys.argv[1]
    return name

def rolldice(remaining):
    """
    Return sum of two simulated dice rolls if sum of remaning numbers is greater than 6,
    and returns one dice roll if sum of remaning numbers if less than 6. 
    """
    a = random.randint(1,6)
    b = random.randint(1,6)
    if score(remaining) > 6:
        return a+b
    else:
        return a
    
def score(a):
    """
    Sums a list of intgers to return to the score of the player 
    at the end of the game.
    
    Inputs: list
    Output: integer
    """
    return sum(a)

def game_over(remaining):
    '''
    End of game procedure.
    '''
    result = score(remaining)
    print("Score for player " + name + " : " + str(result) + " points")
    if result == 0:
        print("Congragulations!! You shut the box!")
    return None
      
name = get_name()
remaining  = [1,2,3,4,5,6,7,8,9]
while True: 
    if score(remaining) != 0:
        roll = rolldice(remaining)
        print("\n")
        print("Numbers left: " + str(remaining))    
        print("Roll: " + str(roll))
        if box.isvalid(roll, remaining):
            while True:
                elim = input("Number to eliminate: ")
                choice = box.parse_input(elim, remaining)
                if len(choice) > 0:
                    if sum(choice) == roll:
                        break
                    else:
                        print("Invalid Input")  
                else:
                    print("Invalid Input")
            remaining = [num for num in remaining if num not in choice]
        else:
            print("Game Over!")
            print("\n")
            game_over(remaining)
            break
    else: 
        game_over(remaining)
        break
```
