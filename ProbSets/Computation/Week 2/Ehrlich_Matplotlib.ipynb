```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
""" Problem 1 """

def mean_row(N):
    sample = np.random.normal(0.0,1.0,(N,N))
    mean = np.mean(sample, axis = 1)
    return mean

def var_row(means):
    var = np.var(means)
    return var


def plot_p1(xdata, ydata):
    plt.plot(xdata, ydata)
    plt.xlabel("Number of Elements Per Row")
    plt.ylabel("Variance")
    plt.title("An Example of the Law of Large Numbers")
    plt.show()

n = np.linspace(100,1000,10)

results = np.zeros_like(n)
for i in range(0, len(n)):
    results[i] = var_row(mean_row(int(n[i]))) 
    
plot_p1(n,results)
```

```python
""" Problem 2 """

def plot_p2(domain):
    
    sin_range = np.sin(domain)
    cos_range = np.cos(domain)
    arctan_range = np.arctan(domain)
    
    plt.plot(domain, sin_range,'r', label = "sine")
    plt.plot(domain, cos_range,'b' , label = "cosine")
    plt.plot(domain, arctan_range,'g', label = "arctan")
    plt.legend(loc='lower right')
    plt.show()
    
domain  = np.linspace(-2*np.pi, 2*np.pi, 1000)

plot_p2(domain)
```

```python
""" Problem 3 """

def plot_p3(domain1, domain2):
    
    fun_range1 = 1/(domain1 - 1)
    fun_range2 = 1/(domain2 - 1)
    
    
    plt.plot(domain1, fun_range1, 'm', lw = 3, linestyle='--')
    plt.plot(domain2, fun_range2, 'm', lw = 3, linestyle='--')
    plt.ylim(-6, 6)
    plt.show()
    
domain1 = np.linspace(-2, 0.999, 500)
domain2 = np.linspace(1.0001, 6, 500)
plot_p3(domain1, domain2)
```

```python
""" Problem 4 """

def plot_p4(ydomain, xdomain):
    
    
    plt.subplot(221)
    plt.plot(xdomain, np.sin(xdomain), 'g')
    plt.ylim(ydomain)
    plt.title("Sin(x)")
    
    plt.subplot(222)
    plt.plot(xdomain, np.sin(2* xdomain),'r', linestyle='--')
    plt.ylim(ydomain)
    plt.title("Sin(2x)")
   
    plt.subplot(223)
    plt.plot(xdomain, 2*np.sin(xdomain),'b', linestyle='--')
    plt.ylim(ydomain)
    plt.title("2Sin(x)")
    
    plt.subplot(224)
    plt.plot(xdomain, 2*np.sin(2*xdomain),'m', linestyle=':')
    plt.ylim(ydomain)
    plt.title("2Sin(2x)")
    
    plt.suptitle("Sine Graphs")
    plt.tight_layout()
    plt.show()
    
ydomain = [-2,2]
xdomain = np.linspace(0, 2*np.pi, 1000)
plot_p4(ydomain, xdomain)
```
